### **인수**
#### **매개변수와 인수**
- **매개변수 vs. 인수**
    - **매개변수** : 함수 내("인터페이스")에서 유효한 변수
    - **인수** : 실제 매개변수에 대입되는 변수
    - 호출시에 **인수가 매개변수에 복제**된다
<br>
#### **인수의 세 가지 종류**
**1. 위치 인수**(positional argument)
    -매개변수 값이 순서에 따라 복제됨
**2. 키워드 인수**(keyword argument)
    -매개변수 키워드에 맞는 변수로 지정 -> 위치 달라도 됨.
**3. 디폴트 인수**(default argument)
    - 매개변수를 정이할 때 값을 미리지정해놓는 것
    -호출 시에 값이 입력되지 않으면 디폴트값 사용.
    - 디폴트 값이 뒤에 와야

```python
# 1. 위치 인수
def menu(wine, entree, dessert):
    return{'wine':wine, 'entree':entree, 'dessert':dessert}

menu('a', 'b', 'c')
```

```python
#2. 키워드 인수
menu(entree='a', dessert='b', wine='c')
```

```python
#3. 디폴트인수
def menu2(wine, entree, dessert='strawberry'): #undefault가 default 앞에 와야
    return{'wine':wine, 'entree':entree, 'dessert':dessert}

print(menu2('rioja', 'egg'))
print(menu2('rioja', 'egg', 'kiwi'))
```

#### **3. 호출방식**
- 정의 부분은 메모리에 저장되어 있다가 호출 시 실행

##### **매개변수 vs. 인수**
- 변수가 넘어가는 게 아니라 변수가 복제가 되는, 참조에 의해 호출되는 것!(call by reference)
- 즉, shallow copy라고 볼 수 있는데, 이 때 문제점이 자료형이 리스트처럼 가변 객체일때 값이 바뀐다는 것이므로(주소값을 공유하기 때문에) 유의!

```python
def multiply(x, y):
    print('1')
    return x * y

print('2')
i = 2
j = 4


print('3')
multiply(i,j) # i = 2, j = 4 -> x = i, y = j(값이 넘어간다) -> shallow copy

print('4')
```

```python
def buggy(arg, result=[]):
    result = [] # shallow copy 되기 때문에 이 빈 리스트를 초기화해줘야 리스트가 쌓이지 않음.
    result.append(arg)
    return (result)


a, b ='a', 'b'
buggy(a)
buggy(b)
```

```python
import string #punct
import nltk

nltk.download('brown')
brown = nltk.corpus.brown
romance = brown.raw(categories='romance')[:5000]
romance
```

#### **문장분절**

```python
# ./. , ? !
# /. => ! 처리 가능함. -> ! 문장에서 제외
# \n
sentences = [sent.strip() for sent in romance.split('./.')]
sentences = [s for s in sentences if s != '!']
sentences[0]
```

```python
[s for s in sentences if '!' in s if s != '!']
```

#### **단어분절**
- sentences-> POS제거, 단어분절

```python
SLASH = '/'
```

```python
example = sentences[0]
[word[:word.find(SLASH)] if word[:word.find(SLASH)] else word for word in example.split()]
```

```python
sentences
```

```python
# 전체 문장에 적용하기
sentence_tokens = [[word[:word.find(SLASH)] \
     if word[:word.find(SLASH)] else word \
         for word in example.split()] for example in sentences]
sentence_tokens
```

#### **punctuation**

```python
punct = list(string.punctuation)
punct
```

```python
# punct 제거하기
[[word for word in sent if word not in punct] for sent in sentence_tokens]
```

```python
#punct에 없는 특수기호 제거하기
my_punct = set([word for sent in sentence_tokens for word in sent if not word[0].isalnum() and word not in punct])
punct.extend(list(my_punct))
```
