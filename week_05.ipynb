## **반복문**

: 주어진 동작을 반복할 때.
- **반복문 종류**
    - **for** : 반복횟수가 정해져서 변하지 않음. 정확한 횟수만큼 실행.
        - 예: 성적 산출
    - **while** : 어떤 조건을 만족하는 동안 반복. 횟수가 정해져있지 않음.
                  if + for 같은 느낌.
                  일정 조건이 거짓이 되면 반복문 빠져나옴.
        - 예: 가위바위보할 때 이겼을 때 끝난다.
        
- **반복문 제어**
    - 사용 지양. 예상치 못한 에러 발생 가능성 있음.
    - **break**: 논리적으로 반복 종료.
    - **else**: 조건이 완전히 끝났을 때 한 번 더 실행. break 시에는 실행 안됨. (어느 지점에서 종료되었는지 확인할 때 사용)
    - **continue**: 어떤 조건 건너뛰고 **다음 조건 실행.**

- for loop 예시

```python
for i in range(10):
    print(i)
```

- while loop 예시

```python
j = 0 # 변수 초기화 블럭 있어야 함
while j < 10:
    print(j)
    j += 1 #변화
```

# **QUIZ - 구구단**

- 변수는 입력 받아서 사용함.
- level 1: n단 출력
- level 2: 1~n단 까지
- level 3: n, m 모두 입력받아서 출력(시작과 끝단)
- level 4: 프로그램으로 만듦.
    - 메뉴: 1) n단 2) 1 ~ n단 3)n~m단 출력
    (사용자가 고르면.)
    구구단 변환기.
    - 메뉴 없는 옵션 선택할 때.

```python
number = int(input("구구단 몇 단을 출력할까요?"))

i = 1
print(f"=== {number}단 ===")
while i <= 9:
    multiply = number * i
    if number > 9:
        break
    else:
        print(f"{number} * {i} = {multiply}")
        i += 1
```

```python
# level 2 풀이
number = int(input("구구단 몇 단까지 출력할까요?"))

i = 1
while i <= number:
    print(f"=== {i}단 ===")
    for j in range(1, 10):
        multiply = i * j
        print(f"{i} * {j} = {multiply}")
    i += 1
```

```python
#Level 3 풀이
while True:
    number_start = int(input("구구단 몇 단부터 출력할까요?"))
    number_end = int(input("구구단 몇 단까지 출력할까요?"))
    if (number_start > 1) & (number_end <= 9) & (number_start <= number_end):
        i = number_start
        while i <= number_end:
            print(f"=== {i}단 ===")
            for j in range(1, 10):
                multiply = i * j
                print(f"{i} * {j} = {multiply}")
            i += 1
        break
    elif number_start < 2:
        print("시작값이 1보다 커야 합니다.")
    elif number_end == 0:
        print("끝값이 9 이하여야 합니다.")
    else:
        print("시작값이 끝값 이하여야 합니다.")
```

```python
#Level 4 풀이
while True:
    menu_num = int(input("메뉴 번호를 고르세요. |  1) n단 2) 1 ~ n단 3) n~m 단 출력"))
    if (1 <= menu_num <=3):
        if menu_num == 1:
            number = int(input("구구단 몇 단을 출력할까요?"))

            i = 1
            print(f"=== {number}단 ===")
            while i <= 9:
                multiply = number * i
                if number > 9:
                    break
                else:
                    print(f"{number} * {i} = {multiply}")
                i += 1
            break

        if menu_num == 2:
            number = int(input("구구단 몇 단까지 출력할까요?"))

            i = 1
            while i <= number:
                print(f"=== {i}단 ===")
                for j in range(1, 10):
                    multiply = i * j
                    print(f"{i} * {j} = {multiply}")
                i += 1
            break
        
        if menu_num == 3:
            while True:
                number_start = int(input("구구단 몇 단부터 출력할까요?"))
                number_end = int(input("구구단 몇 단까지 출력할까요?"))
                if (number_start > 1) & (number_end <= 9) & (number_start <= number_end):
                    i = number_start
                    while i <= number_end:
                        print(f"=== {i}단 ===")
                        for j in range(1, 10):
                            multiply = i * j
                            print(f"{i} * {j} = {multiply}")
                        i += 1
                    break
                elif number_start < 2:
                    print("시작값이 1보다 커야 합니다.")
                elif number_end == 0:
                    print("끝값이 9 이하여야 합니다.")
                else:
                    print("시작값이 끝값 이하여야 합니다.")
        break
    else:
        print("메뉴 번호는 1 ~ 3 중에서 골라야 합니다.")
```

### Brown 활용하기 
#### - conditional frequency distribution(조건에 따른 fdist)

**- categories()**

```python
brown = nltk.corpus.brown
```

```python
brown.categories()
```

```python
fdist = nltk.FreqDist([w.lower() for w in brown.words(categories='adventure')])
```

```python
fdist['men']
```

```python
# modal verbs
modals = ['can', 'could', 'will', 'would', 'may', 'might']
          
for verb in modals:
    print(verb, fdist[verb])
```

```python
cfd = nltk.ConditionalFreqDist(# 각 장르에 따른 frequency distribution을 만든다.
    (genre, word.lower())
    for genre in brown.categories()
    for word in brown.words(categories=genre)
)
```
